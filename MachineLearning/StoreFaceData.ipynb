```python
import cv2
import numpy as np
```

```python
dataset = cv2.CascadeClassifier('data.xml')
```

```python
face_data = []
video = cv2.VideoCapture(0)
while True:
    flag, img = video.read()
    img = cv2.resize(img,None,fx=0.5,fy=0.5)
    if flag:
        faces = dataset.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,100,255),4)
            
            face = img[y:y+h,x:x+w,:]
            face = cv2.resize(face,(50,50))
            if len(face_data) < 200:
                face_data.append(face)
                print(len(face_data))
            
        cv2.imshow('result',img)
        if cv2.waitKey(10) == ord('q') or len(face_data) >= 200:
            break
    else:
        print("Camera not working")

face_data = np.asarray(face_data)
np.save('my_face.npy',face_data)

video.release()
cv2.destroyAllWindows()
```
