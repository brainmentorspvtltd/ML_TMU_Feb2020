```python
import urllib.request as url
import bs4
```

```python
for i in range(1,5):
    path = f"https://www.flipkart.com/search?q=tv&page={i}"
    res = url.urlopen(path)
    page = bs4.BeautifulSoup(res)
    titles = page.findAll('div',class_='_3wU53n')
    priceList = page.findAll('div',class_='_1vC4OE _2rQ-NK')
    for j in range(len(titles)):
        print("Title",titles[i].text)
        print("Price",priceList[i].text)
        print("="*20)
```
